## Linked List 

```python
class node:
    def __init__(self,data=None):
        self.data=data
        self.next=None

class linked_list:
    def __init__(self):
        # the head node is not going to contain any data it will only used to referene the first node address
        self.head=node() 
        self.length=0;
        self.tail = self.head
        

    # Adds new node containing 'data' to the end of the linked list.
    def append(self,data):
        new_node=node(data)
        cur=self.head
        while cur.next!=None:
            cur=cur.next
        cur.next=new_node
        self.length += 1
        self.tail.next = new_node
        self.tail = new_node
        
    def prepend(self,data):
        new_node = node(data);
        temp_node = self.head;
        new_node.next = temp_node.next
        self.head.next = new_node
        self.length += 1
         

    # Returns the value of the node at 'index'.
    def traverseToIndex(self,index):  
        if(index >= self.length):
            print('index out of bound')
            return None
        cur_node = self.head
        curr_index = 0
        while cur_node.next!=None:
            cur_node=cur_node.next
            if(curr_index == index): return cur_node.data
            curr_index += 1

    # Remove node 
    def remove(self,index):
        if(index >= self.length):
            print('index out of bound')
            return None
        cur_node = self.head
        curr_index = 0
        while cur_node.next!=None:
            cur_node=cur_node.next
            if(curr_index == index-1): 
                leader = cur_node
                break;
            curr_index += 1
        
        unwantedNode = leader.next
        leader.next = unwantedNode.next
        self.length -=1

    def insertToIndex(self,index,data):
        if(index >= self.length):
            print('index out of bound')
            return None
        cur_node = self.head
        curr_index = 0
        while cur_node.next!=None:
            cur_node=cur_node.next
            if(curr_index == index-1): 
                leader = cur_node
                break;
            curr_index += 1
        new_node=node(data)
        holdingIndex = leader.next;
        leader.next = new_node
        new_node.next = holdingIndex
        self.length +=1


    # Returns the length (integer) of the linked list.
    def getlength(self):
        cur=self.head
        total=0
        while cur.next!=None:
            total+=1
            cur=cur.next
        return total 

    # Prints out the linked list in traditional Python list format. 
    def display(self):
        elems=[]
        cur_node=self.head
        while cur_node.next!=None:
            cur_node=cur_node.next
            elems.append(cur_node.data)
        print(elems)

```

```python
"""
1290. Convert Binary Number in a Linked List to Integer

Given head which is a reference node to a singly-linked list. The value of each node in the linked list is either 0 or 1. The linked list holds the binary representation of a number.
Return the decimal value of the number in the linked list.

Input: head = [1,0,1]
Output: 5
Explanation: (101) in base 2 = (5) in base 10
"""

# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def getDecimalValue(self, head: ListNode) -> int:
        numList =[]
        curr = head 
        while curr != None:
            numList.append(curr.val)
            curr = curr.next
        # print(numList)
        decimalNum = 0
        decimalPower = 0
        for val in reversed(numList):
            if(val):
                decimalNum += 2**decimalPower
            decimalPower +=1
        return decimalNum


# driver code 
mainNode = ListNode(0);
numList = [1,0,1]
curr = mainNode;
for i in numList:
    newNode = ListNode(i)
    curr.next = newNode;
    curr = curr.next;

sol = Solution()
print(sol.getDecimalValue(mainNode.next))
```

```python
"""
876. Middle of the Linked List
Given a non-empty, singly linked list with head node head, return a middle node of linked list.
If there are two middle nodes, return the second middle node.

Input: [1,2,3,4,5]
Output: Node 3 from this list (Serialization: [3,4,5])
The returned node has value 3.  (The judge's serialization of this node is [3,4,5]).
Note that we returned a ListNode object ans, such that:
ans.val = 3, ans.next.val = 4, ans.next.next.val = 5, and ans.next.next.next = NULL.
"""
# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def middleNode(self, head: ListNode) -> ListNode:
        curr = head;
        nodeCount = 0
        while curr != None:
            nodeCount += 1
            curr = curr.next
            
        middle = nodeCount//2
        currNodeCount = 0
        curr = head;
        while curr != None:
            if currNodeCount == middle:
                head = curr
                return head
            curr = curr.next
            currNodeCount+=1
            
# driver code 
mainNode = ListNode(0);
numList = [1,2,3,4,5]
curr = mainNode;
for i in numList:
    newNode = ListNode(i)
    curr.next = newNode;
    curr = curr.next;

sol = Solution()
node=sol.middleNode(mainNode.next)

while node != None:
    print(node.val,end=" ")
    node = node.next
```

```python
# Definition for singly-linked list.
"""
328. Odd Even Linked List

Given a singly linked list, group all odd nodes together followed by the even nodes. 
Please note here we are talking about the node number and not the value in the nodes.
You should try to do it in place. The program should run in O(1) space complexity and O(nodes) time complexity.

Input: 1->2->3->4->5->NULL
Output: 1->3->5->2->4->NULL

Input: 2->1->3->5->6->4->7->NULL
Output: 2->3->6->7->1->5->4->NULL

Note:
The relative order inside both the even and odd groups should remain as it was in the input.
The first node is considered odd, the second node even and so on ...
"""
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def oddEvenList(self, head: ListNode) -> ListNode:
        if head == None :
            return head
        oddList = []
        evenList = []
        curr = head;
        nodeCount = 1
        while curr != None:
            if nodeCount & 1 :
                oddList.append(curr.val)
            else:
                evenList.append(curr.val)
            curr = curr.next
            nodeCount += 1
#         print(oddList)
#         print(evenList)
        finalNodeList = oddList + evenList 
        # print(finalNodeList)
        head = ListNode(finalNodeList[0])
        curr = head
        
        for i in range(1,len(finalNodeList)):
            newNode = ListNode(finalNodeList[i])
            curr.next = newNode;
            curr = curr.next;
        return head
            
            
# driver code 
mainNode = ListNode(0);
numList = [2,1,3,5,6,4,7]
curr = mainNode;
for i in numList:
    newNode = ListNode(i)
    curr.next = newNode;
    curr = curr.next;

sol = Solution()
result = sol.oddEvenList(mainNode.next)
curr = result 
while curr != None:
    print(curr.val,end="->");
    if curr.next == None:
        print('NULL')
    curr = curr.next
```

```python
"""
206. Reverse Linked List
Reverse a singly linked list.

Input: 1->2->3->4->5->NULL
Output: 5->4->3->2->1->NULL

"""

# Definition for singly-linked list.
class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def reverseList(self, head: ListNode) -> ListNode:
        prev = None
        current = head 
        while(current is not None): 
            next = current.next
            current.next = prev 
            prev = current 
            current = next
        head = prev
        return head
        

# driver code 
mainNode = ListNode(0);
numList = [2,1,3,5,6,4,7]
curr = mainNode;
for i in numList:
    newNode = ListNode(i)
    curr.next = newNode;
    curr = curr.next;

sol = Solution()
result = sol.reverseList(mainNode.next)
curr = result 
while curr != None:
    print(curr.val,end="->");
    if curr.next == None:
        print('NULL')
    curr = curr.next
```
