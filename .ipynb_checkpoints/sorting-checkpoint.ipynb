```python
def selectionSort(nums):
    for i in range(len(nums)):
        imin = i
        for j in range(i + 1,len(nums)):
            if (nums[j] < nums[imin]):
                imin = j;
        
        temp = nums[i];
        nums[i] = nums[imin];
        nums[imin] = temp;
    print(nums)

selectionSort([-2, 1, 4, 6, 89, 9])
```
